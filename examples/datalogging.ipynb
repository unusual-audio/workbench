```python
from matplotlib import pyplot as plt
import numpy as np

from workbench.datalogging.csv import CSVDataLogger

with CSVDataLogger('measurements.csv') as logger:
    logger.log("test-measurement", 0.11, "example1")
    logger.log("test-measurement", 0.12, "example1")
    logger.log("test-measurement", 0.13, "example1")

    r = logger.query(np.datetime64("2023-01-01T00:00:00"), experiment="example1")

plt.plot(r["time"], r["test-measurement"])
plt.show()
```
